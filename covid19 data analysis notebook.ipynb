### task 3.2: find maxmimum infection rate for China

```python
cdata_agg.loc['India'].diff().max()
```

```python
cdata_agg.loc['China'].diff().max()
```

### Task 3.3: find maximum infection rate for all of the countries. 

```python
countries=list(cdata_agg.index)
max_infection_rate=[]
for c in countries:
    max_infection_rate.append(cdata_agg.loc[c].diff().max())
cdata_agg['max_infection_rate']=max_infection_rate    
```

```python
cdata_agg.head()
```

### Task 3.4: create a new dataframe with only needed column 

```python
corod=pd.DataFrame(cdata_agg['max_infection_rate'])
```
